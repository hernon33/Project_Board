```python
import pandas as pd

file_path = "/content/drive/MyDrive/Colab Notebooks/DATA6520/hr_data.csv"
df = pd.read_csv(file_path)

dept_turnover = (
    df.groupby("Department")
      .agg(
          total_employees=("left", "count"),
          employees_left=("left", "sum")
      )
)

dept_turnover["turnover_rate"] = (
    (dept_turnover["employees_left"] / dept_turnover["total_employees"] * 100)
    .round(2)
    .astype(str) + "%"
)

dept_turnover = dept_turnover.sort_values("turnover_rate", ascending=False)

print(dept_turnover)
```

```python
import pandas as pd

file_path = "/content/drive/MyDrive/Colab Notebooks/DATA6520/hr_data.csv"
df = pd.read_csv(file_path)

def tenure_band(years):
    if years <= 3:
        return "early_tenure"
    elif years <= 6:
        return "mid_tenure"
    else:
        return "long_tenure"

df["tenure_band"] = df["time_spend_company"].apply(tenure_band)

tenure_turnover = (
    df.groupby("tenure_band")
      .agg(
          total_employees=("left", "count"),
          employees_left=("left", "sum")
      )
)

tenure_turnover["turnover_rate"] = (
    (tenure_turnover["employees_left"] / tenure_turnover["total_employees"] * 100)
    .round(2)
    .astype(str) + "%"
)

print(tenure_turnover)
```

```python
import pandas as pd

file_path = "/content/drive/MyDrive/Colab Notebooks/DATA6520/hr_data.csv"
df = pd.read_csv(file_path)

def workload_band(hours):
    if hours < 160:
        return "low_workload"
    elif hours <= 220:
        return "moderate_workload"
    else:
        return "high_workload"

df["workload_level"] = df["average_montly_hours"].apply(workload_band)

workload_turnover = (
    df.groupby("workload_level")
      .agg(
          total_employees=("left", "count"),
          employees_left=("left", "sum")
      )
)

workload_turnover["turnover_rate"] = (
    (workload_turnover["employees_left"] / workload_turnover["total_employees"] * 100)
    .round(2)
    .astype(str) + "%"
)

print(workload_turnover)
```
